```python
#This is a children's game,where the player is given 3 chances to input,s for scissor, r for rock and p for paper
#and for each input computer will generate a opposing random output, comparison of these two in three will decide winner 
import random , sys
wins=0
loses=0
ties=0
GMLIMIT=3 #variable for 3 no of chances. can be modified to any number of chances
chances=0 #counter for counting nd limiting chances
print('ROCK PAPER SCISSORS')
while chances<=GMLIMIT: #main loop for the game
    print('%s Wins,%s Losses, %s ties' % (wins,loses,ties)) #header 
    while True and chances<=GMLIMIT: #loop for each input
            print('make your move r,s,p or q')
            playermove=input() #User input
            
            #randomnumber=random.randint(1,3)
            if playermove=='q':
                chances=GMLIMIT
                continue
            elif playermove=='s' or playermove=='r' or playermove=='p': #validation of input
                
                break
            else:
                print('type one of r,s,p or q')  #direction for erronous input    
    chances=chances+1 #increamenting counter       
    if chances==GMLIMIT:
        break
    if playermove=='r':  
     print('ROCK VERSES ...')    #display of players input
    elif playermove=='p':
     print('PAPER VERSES ...') 
    elif playermove=='s':
     print('SCISSORS VERSES ...')
    randomnumber=random.randint(1,3) #generation of computer random output
    if  randomnumber ==1:
      computermove = 'r'
      print('ROCK')            #display of computer output
    elif  randomnumber ==2:
      computermove = 'p'
      print('PAPER')        
    elif  randomnumber ==3:
      computermove = 's'
      print('SCISSOR')
    if playermove== computermove: #logic for comparison in next of the lines counting wins and losses.
        print('It is a Tie')
        ties=ties+1
        continue
    elif playermove=='r' and computermove=='s':
        print('You Win')
        wins=wins+1
        continue
    elif playermove=='p' and computermove=='r':
        print('You Win')
        wins=wins+1
        continue
    elif playermove=='s' and computermove=='p':
        print('You Win')
        wins=wins+1
        continue
    elif playermove=='r' and computermove=='p':
        print('Your Loss')
        loses=loses+1
        continue
    elif playermove=='p' and computermove=='s':
        print('Your Loss')
        loses=loses+1
        continue
    elif playermove=='s' and computermove=='r':
        print('Your Loss')
        loses=loses+1 
        continue
    if playermove=='q':
        print('YOU QUIT')
        continue
if wins<loses:
    print('YOU LOSE')   #declaration of final result.
else:
    print('YOU WIN')
```
